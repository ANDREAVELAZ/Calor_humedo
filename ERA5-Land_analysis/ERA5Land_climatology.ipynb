```python
import xarray as xr
import cartopy
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np
```

```python
#Función para la serie de tiempo
def serieT_mean(variable,unidades=0):
    ar=variable.resample(time="YE").mean("time")  
    ar=ar.mean("time")-unidades
    return ar
```

```python
#Función para la serie de tiempo
def serieT_max(variable,unidades=0):
    ar=variable.resample(time="YE").max("time")  
    ar=ar.max("time")-unidades
    return ar
```

```python
#Variables de temperatura
Tmax=xr.open_dataset("T_max.nc")
Tmean=xr.open_dataset("T_mean.nc")

#Variables de humedad
qmax=xr.open_dataset("q_max.nc")
qmean=xr.open_dataset("q_mean.nc")

#Variables de TW
Twmax=xr.open_dataset("TWB_dmax.nc")
Twmean=xr.open_dataset("TWB_dmean.nc")
```

```python
#Aplicando la función de la serie temporal
Tmax_c=serieT_max(Tmax.T_max)
Tmean_c=serieT_mean(Tmean.T_mean)

qmax_c=serieT_max(qmax.q_max)
qmean_c=serieT_mean(qmean.q_mean)

Twmax_c=serieT_max(Twmax.TWB_dmax)
Twmean_c=serieT_mean(Twmean.TWB_dmean)
```

```python
datos=[Tmean_c,qmean_c*1000,Twmean_c,Tmax_c,qmax_c*1000,Twmax_c]
colors=["YlOrRd","PuBu","Reds","YlOrRd","PuBu","Reds"]
name=["[$^{\circ}$C]","[$g$ $kg^{-1}$]"," [$^{\circ}$C]"," [$^{\circ}$C]"," [$g$ $kg^{-1}$]"," [$^{\circ}$C]"]
inc=["a) $T_{mean}$","b) $SH_{mean}$","c) $T_{W,mean}$","d) $T_{maxr}$","e) $SH_{maxr}$","f) $T_{W,maxr}$"]
```

```python
fig = plt.figure(figsize=(10, 7),dpi=300)

for idata,data in enumerate(datos):
    
    models=data
    
    ax = fig.add_subplot(2, 3, 1+idata, projection=ccrs.PlateCarree())
    # agregamos lineas costeras
    ax.coastlines()
    ax.add_feature(cfeature.BORDERS)
    if idata==5:
        levelss=np.arange(15,32,2)
        cs=ax.contourf(data.longitude,data.latitude,models,levels=levelss,cmap=colors[idata],extend='both',transform=ccrs.PlateCarree())
    
    else:
        cs=ax.contourf(data.longitude,data.latitude,models,cmap=colors[idata],extend='both',transform=ccrs.PlateCarree())
    
    gl=ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,color='gray', alpha=0.5, linestyle='--')
    ax.xlabels_top = False# It should turn off the upper x ticks
    gl.top_labels = gl.right_labels = False
    
    if idata!=0 and idata!=3:
        gl.left_labels=False
    
    cbar=plt.colorbar(cs,orientation='horizontal',fraction=0.1,pad=0.09)
    cbar.ax.set_xlabel(name[idata])
    
    plt.text(-112, 15, inc[idata], fontsize=13, color="black", ha="center", va="center", bbox=dict(facecolor="lightgray", edgecolor="black", boxstyle="round",pad=0.4))
    
plt.tight_layout(h_pad=0.5,pad=0.9)
plt.subplots_adjust(top=0.95)

plt.savefig("ERA_Map.svg",bbox_inches="tight",dpi=300)
plt.savefig("ERA_Map.png",bbox_inches="tight",dpi=300)
```
