```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pandas as pd
from sklearn.cluster import MeanShift
# Import CMasher to register colormaps
import cmasher as cmr
import geopandas as gpd
import cftime
from scipy.stats import linregress
```

```python
df=pd.read_csv("IDD.csv",header=None)
ID=df[(df[0]>="760011-99999")&(df[0]<="780160-13601")] 
ID1=ID[0].str.split(pat="-", expand=True) #Separando 
ID["ID"]=ID1[0].astype(int)

esta=pd.read_fwf("ID.txt",header=None)
p=esta[1]
filtered_df = ID[(ID["ID"]).isin(p)]
filtered_df #Encontrando los datos de las estaciones que ocupamos
lat=filtered_df[2].values
lon=filtered_df[3].values
ele=filtered_df[4].values

#Agregando la elevación al dataframe
esta["elevacion"]=ele
esta["Lat"]=lat
esta["Lon"]=lon
esta["Lugar"]=filtered_df[1].values
```

```python
#Cargando los datos con los que se harán los clusters
da_list=[]
for i,ii in enumerate(esta[1]):
    datos=xr.open_dataset(f"/home/jovyan/Tesis/HadISD/estadísticos/{ii}99999_estadisticos.nc")
    datos=datos.assign_coords(estacion=ii)
    datos=datos.expand_dims('estacion')
    da_list.append(datos)
seas=xr.concat(da_list,dim='estacion')
```

```python
#Seleccionando los datos que van de 1979 en adelante
recorte_seas=seas.where(seas.time.dt.year>1978,drop=True)

#Sacando la serie de tiempo
serie=(recorte_seas.resample(time="ME").mean()).resample(time="YE").mean()
serieT=serie.mean("estacion")
fig = plt.figure(figsize=(15,9),dpi=300)
datos=["a) $T_{W,min}$","b) $ T_{W,max}$","c) $T_{W,mean}$","d) $T_{min}$","e) $T_{max}$","f) $T_{mean}$","g) $SH_{min}$","h) $SH_{max}$","i) $SH_{mean}$"]
#variables=["$Tw_{min}$","$Tw_{max}$","$Tw_{mean}$","$T_{min}$","$T_{max}$","$T_{mean}$","$SH_{min}$","$SH_{max}$","$SH_{mean}$"]
#labels=["$TW_{min}$ $[^{\circ}C]$","$TW_{max}$ $[^{\circ}C]$","$TW_{mean}$ $[^{\circ}C]$"," $T_{min}$ $[^{\circ}C]$","$T_{max}$ $[^{\circ}C]$","$T_{mean}$ $[^{\circ}C]$","$SH_{min}$ $[g/kg]$","$SH_{max}$ $[g/kg]$","$SH_{mean}$ $[g/kg]$"]
labels=["$[^{\circ}C]$","$[^{\circ}C]$","$[^{\circ}C]$","$[^{\circ}C]$"," $[^{\circ}C]$","$[^{\circ}C]$","[$g$ $kg^{-1}$]","[$g$ $kg^{-1}$]"," [$g$ $kg^{-1}$]"]
n=1
for i,ii in enumerate (np.array(serieT.variables)):
    from decimal import Decimal, ROUND_DOWN

   
    if ii=="Twmax" or ii=="Twmin" or ii=="Twmean":
        x=serieT[ii].time.dt.year
        y=serieT[ii]-273.15
       
        plt.subplot(3,3,n)
      
        slope, intercept, r_value, p_value, std_err = linregress(x, y)
        slope1=Decimal(float(slope)).quantize(Decimal("0.01"), rounding=ROUND_DOWN)
       
        plt.plot(x, slope*x + intercept, color="black", label=fr"$\beta_{1}$={slope1*10} $^\circ$C $dec^{-1}$")
        plt.plot(serieT.time.dt.year,serieT[ii]-273.15,color="orange")
        plt.grid(linestyle="--", color="grey",alpha=0.6)
        plt.xticks(color="white",rotation=0)
        plt.legend(fontsize=8)

        n=n+1
        if ii=="Twmin":
            plt.ylabel(labels[i])
            plt.text(2016, 14.47, datos[i], fontsize=12, color="black",
            ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
        elif  ii=="Twmax" or ii=="Twmean":
            if ii=="Twmax":    
                plt.text(2015, 18.58, datos[i], fontsize=12, color="black",
                ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
                plt.yticks(np.arange(18.4,20.26,0.3))
                
            else:
                plt.text(2015, 16.6, datos[i], fontsize=12, color="black",
                ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
                plt.yticks(np.arange(16.4,18.2,0.3))
    elif ii=="SHmax" or ii=="SHmin" or ii=="SHmean":
        x=serieT[ii].time.dt.year
        y=serieT[ii]*1000
        plt.subplot(3,3,n)
        slope, intercept, r_value, p_value, std_err = linregress(x, y)
        slope1=Decimal(float(slope)).quantize(Decimal("0.001"), rounding=ROUND_DOWN)
    
        plt.plot(x, slope*x + intercept, color="black", label=fr"$\beta_{1}$={slope1*10} $g$ $kg^{-1}$ $dec^{-1}$")
        plt.plot(serieT.time.dt.year,serieT[ii]*1000,color="blue")
      
        plt.grid(linestyle="--", color="grey",alpha=0.6)
        plt.legend(fontsize=8)
        plt.xticks(rotation=0)
        
        n=n+1
        if ii=="SHmin":
           plt.ylabel(labels[i]) 
           plt.text(2016, 10.23, datos[i], fontsize=12, color="black",
            ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
           plt.yticks(np.arange(10.2,11.1,0.2))
        elif  ii=="SHmax" or ii=="SHmean":
            if ii=="SHmax":    
                plt.text(2015.5, 13.09, datos[i], fontsize=12, color="black",
                ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
            else:
                plt.text(2015.3, 11.79, datos[i], fontsize=12, color="black",
                ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
            
    elif ii=="Tmax" or ii=="Tmin" or ii=="Tmean":
        x=serieT[ii].time.dt.year
        y=serieT[ii]
        plt.subplot(3,3,n)
        slope, intercept, r_value, p_value, std_err = linregress(x, y)
        slope1=Decimal(float(slope)).quantize(Decimal("0.01"), rounding=ROUND_DOWN)
        plt.plot(x, slope*x + intercept, color="black", label=fr"$\beta_{1}$={slope1*10} $^\circ$C $dec^{-1}$")
        plt.plot(serieT.time.dt.year,serieT[ii],color="red")
       
        plt.grid(linestyle="--", color="grey",alpha=0.6)
        plt.legend(fontsize=8)
        plt.xticks(color="white",rotation=0)
   
        
        n=n+1
        if ii=="Tmin":
            plt.ylabel(labels[i])
            plt.text(2017.5, 16.8, datos[i], fontsize=12, color="black",
            ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
            plt.yticks(np.arange(16.6,18.6,0.3))
        elif  ii=="Tmax" or ii=="Tmean":
            if ii=="Tmax":    
                plt.text(2017.1, 25.55, datos[i], fontsize=12, color="black",
                ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))
                
            else:
                plt.text(2016.7, 21.2, datos[i], fontsize=12, color="black",
                ha="left", va="top", bbox=dict(facecolor="lightgray", edgecolor="black",
                                           boxstyle="round", pad=0.4))

    plt.tight_layout(h_pad=-0.9,pad=0.5,w_pad=0.8)
plt.savefig("serietemporal_all.svg",dpi=300,bbox_inches="tight")
plt.savefig("serietemporal_all.png",dpi=300,bbox_inches="tight")
```
