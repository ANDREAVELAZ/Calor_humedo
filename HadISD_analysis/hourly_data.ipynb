```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
import xarray as xr
```

```python
#Cargando archivo con las id de las estaciones HadISD en México
esta=pd.read_fwf("/home/andreagvc/data/ERA/ERA/ID.txt",header=None)
```

```python
#Concatenando los datos horarios en un solo archivo
da_list=[]
for i in esta[1]:
    datos=xr.open_dataset(f"/home/andreagvc/data/ERA/ERA/Datos_HAD/{i}99999.nc") #Ruta de los datos horarios
    datos=datos.assign_coords(estacion=i)
    datos=datos.expand_dims('estacion')
    da_list.append(datos)
#print(len(da_list))
estda=xr.concat(da_list,dim='estacion')
  
```

```python
#Seleccionando Tw para conocer la cantidad de datos por hora
Tw=estda.Tw

#Promediando las estaciones para quedarnos con la dimensión temporal
Tw_time=Tw.mean("estacion")
```

```python
#Agrupando y contando los datos por hora
hourly_data=Tw_time.groupby("time.hour").count()
```

```python
plt.scatter(hourly_data.hour,hourly_data)
plt.xticks(np.arange(0, 24, 2))
plt.xlabel("Hora")
plt.grid(color='gray', alpha=0.5, linestyle='--')
plt.ylabel("Núm. de observaciones totales")
plt.show()

plt.savefig("hourly_data_obs.svg",dpi=150,bbox_inches="tight")
plt.savefig("hourly_data_obs.jpg",dpi=150,bbox_inches="tight")
```
