```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pandas as pd
import cmasher as cmr
import datetime
import cftime
import glob
```

```python
#Función para la serie de tiempo de los máximos
def serieT_max(variable,unidades=0):
    ar=variable.resample(time="YE").max("time")  
    ar=ar.max("time")-unidades
    return ar
```

```python
#Función para la serie de tiempo de los promedios
def serieT_mean(variable,unidades=0):
    ar=variable.resample(time="YE").mean("time")  
    ar=ar.mean("time")-unidades
    return ar
```

```python
#Variables de temperatura
Tmax=xr.open_dataset("/home/andreagvc/data/ERA/ERA/T_max.nc")
Tmean=xr.open_dataset("/home/andreagvc/data/ERA/ERA/T_mean.nc")

#Variables de humedad
qmax=xr.open_dataset("/home/andreagvc/data/ERA/ERA/q_max.nc")
qmean=xr.open_dataset("/home/andreagvc/data/ERA/ERA/q_mean.nc")

#Variables de TW
Twmax=xr.open_dataset("/home/andreagvc/data/ERA/ERA/TWB_dmax.nc")
Twmean=xr.open_dataset("/home/andreagvc/data/ERA/ERA/TWB_dmean.nc")
```

```python
#Cargando las fechas de la canícula
MSD=pd.read_csv("average_dates_1979-2018.csv")

#PASANDO LAS FECHAS A FORMATO DATETIME
MSD["Onset"]=pd.to_datetime(MSD["Onset"])
MSD["Year"]=pd.to_datetime(MSD["Year"])
MSD["Retreat"]=pd.to_datetime(MSD["Retreat"])
MSD["MSDO"]=pd.to_datetime(MSD["MSDO"])
MSD["MSDE"]=pd.to_datetime(MSD["MSDE"])
```

```python
anio = pd.to_datetime([f"{año}-03-01" for año in range(1979, 2019) if año not in [1987]])
anio1 = pd.to_datetime([f"{año}-12-31" for año in range(1979, 2019) if año not in [1987]])
```

```python
lista_datetime = [fecha.to_pydatetime() for fecha in anio]
lista_datetime1 = [fecha.to_pydatetime() for fecha in anio1]
```

```python
MSD["DS"]=lista_datetime
MSD["DW"]=lista_datetime1
```

```python
#Recorte Early Dry Season
def DS_rec(datos,MSD):
    recortes = []

# Iterar sobre cada año en el CSV
    for i, row in MSD.iterrows():
        inicio =row["DS"]
        fin = row["Onset"]

        # Recortar usando slice
        recorte = datos.sel(time=slice(inicio, fin))
        recortes.append(recorte)

    # Unir todos los recortes en un solo Dataset
    dry_spring= xr.concat(recortes, dim="time")
    return dry_spring
```

```python
#Máximos
Tw_dry_springmax=DS_rec(Twmax,MSD)
T_dry_springmax=DS_rec(Tmax,MSD)
q_dry_springmax=DS_rec(qmax,MSD)
```

```python
#Promedios
Tw_dry_springmean=DS_rec(Twmean,MSD)
T_dry_springmean=DS_rec(Tmean,MSD)
q_dry_springmean=DS_rec(qmean,MSD)
```

```python
#Midsummer Drought
def MSD_period(datos,MSD):
    recortes = []

# Iterar sobre cada año en el CSV
    for i, row in MSD.iterrows():
        inicio =row["MSDO"]
        fin = row["MSDE"]

        # Recortar usando slice
        recorte = datos.sel(time=slice(inicio, fin))
        recortes.append(recorte)
    g= xr.concat(recortes, dim="time")

    return g
```

```python
#Máximos
Tw_MSD_max=MSD_period(Twmax,MSD)
T_MSD_max=MSD_period(Tmax,MSD)
q_MSD_max=MSD_period(qmax,MSD)
```

```python
#Promedios
Tw_MSD_mean=MSD_period(Twmean,MSD)
T_MSD_mean=MSD_period(Tmean,MSD)
q_MSD_mean=MSD_period(qmean,MSD)
```

```python
#Deep wet season
def MSD_winterdry (datos,MSD):
    recortes = []

# Iterar sobre cada año en el CSV
    for i, row in MSD.iterrows():
        inicio =row["Retreat"]
        fin = row["DW"]

        # Recortar usando slice
        recorte = datos.sel(time=slice(inicio, fin))
        recortes.append(recorte)
    g= xr.concat(recortes, dim="time")
    
    return g
```

```python
#Máximos
Tw_MSD_wdry_max=MSD_winterdry(Twmax,MSD)
T_MSD_wdry_max=MSD_winterdry(Tmax,MSD)
q_MSD_wdry_max=MSD_winterdry(qmax,MSD)
```

```python
#Promedios
Tw_MSD_wdry_mean=MSD_winterdry(Twmean,MSD)
T_MSD_wdry_mean=MSD_winterdry(Tmean,MSD)
q_MSD_wdry_mean=MSD_winterdry(qmean,MSD)
```

```python
#First Peak
def onset_MSDO (datos,MSD):
    recortes = []

# Iterar sobre cada año en el CSV
    for i, row in MSD.iterrows():
        inicio =row["Onset"]
        fin = row["MSDO"]

        # Recortar usando slice
        recorte = datos.sel(time=slice(inicio, fin))
        recortes.append(recorte)
    g= xr.concat(recortes, dim="time")
    
    return g
```

```python
Tw_fpeak_ERA_max=onset_MSDO(Twmax,MSD)
T_fpeak_ERA_max=onset_MSDO(Tmax,MSD)
q_fpeak_ERA_max=onset_MSDO(qmax,MSD)
```

```python
Tw_fpeak_ERA_mean=onset_MSDO(Twmean,MSD)
T_fpeak_ERA_mean=onset_MSDO(Tmean,MSD)
q_fpeak_ERA_mean=onset_MSDO(qmean,MSD)
```

```python
#Second Peak
def MSDE_retreat (datos,MSD):
    recortes = []

# Iterar sobre cada año en el CSV
    for i, row in MSD.iterrows():
        inicio =row["MSDE"]
        fin = row["Retreat"]

        # Recortar usando slice
        recorte = datos.sel(time=slice(inicio, fin))
        recortes.append(recorte)
    g= xr.concat(recortes, dim="time")
    return g
```

```python
#Máximos
Tw_speak_ERA_max=MSDE_retreat(Twmax,MSD)
T_speak_ERA_max=MSDE_retreat(Tmax,MSD)
q_speak_ERA_max=MSDE_retreat(qmax,MSD)
```

```python
#Promedios
Tw_speak_ERA_mean=MSDE_retreat(Twmean,MSD)
T_speak_ERA_mean=MSDE_retreat(Tmean,MSD)
q_speak_ERA_mean=MSDE_retreat(qmean,MSD)
```

```python
datos=[Tw_dry_springmax.TWB_dmax.max("time"),Tw_fpeak_ERA_max.TWB_dmax.max("time"),Tw_MSD_max.TWB_dmax.max("time"),Tw_speak_ERA_max.TWB_dmax.max("time"),Tw_MSD_wdry_max.TWB_dmax.max("time")]
colors=["YlOrRd","YlOrRd","YlOrRd","YlOrRd","YlOrRd"]
name=["[$^{\circ}$C]"," [$^{\circ}$C]"," [$^{\circ}$C]"," [$^{\circ}$C]"," [$^{\circ}$C]",]
inc=["a) $DS$","b) $FP$","c) $MSD$","d) $SP$","e) $DW$"]
```

```python
def add_map_colorbar(pos,axis,labelsita,fig,cs):
        # posicion pos
        # axis 
        # label del coorbar
        # objeto de figura (fig)
        # resultado o barra colorear
        cbar_ax = fig.add_axes(axis)
        cbar=fig.colorbar(cs, cax=cbar_ax,orientation=pos,shrink=11)
        cbar.ax.tick_params(labelsize=8)
        if labelsita!=None:
                cbar.set_label(label=labelsita,fontsize=12)
        return cbar
```

```python
fig = plt.figure(figsize=(10, 7),dpi=300)

for idata,data in enumerate(datos):
   
    models=data
    
    ax = fig.add_subplot(2, 3, 1+idata, projection=ccrs.PlateCarree())
    # agregamos lineas costeras
    ax.coastlines()
    ax.add_feature(cfeature.BORDERS)
    levelss=np.arange(15,32,0.5)
    cs=ax.contourf(data.longitude,data.latitude,models,levels=levelss,cmap=colors[idata],extend='both',transform=ccrs.PlateCarree())
    
    ax.set_extent([-120,-86, 12,34], crs=ccrs.PlateCarree())
    # lineas de paralelos
    gl=ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,color='gray', alpha=0.5, linestyle='--')
    ax.xlabels_top = False# It should turn off the upper x ticks
    gl.top_labels = gl.right_labels = False
    
    if idata!=0 and idata!=3:
        gl.left_labels= False
    if idata<2:
        gl.bottom_labels= False
         
    else:   
        gl.bottom_labels= True
  
    plt.text(-115, 15, inc[idata], fontsize=10, color="black", ha="center", va="center", bbox=dict(facecolor="lightgray", edgecolor="black", boxstyle="round",pad=0.4)
            ,transform=ccrs.PlateCarree())
    
    add_map_colorbar('horizontal',[0.12,0.1,0.8,0.03],"[$^\circ$C]",fig,cs)
    
plt.suptitle(r"$\mathbf{Máximos}$ $\mathbf{de}$ $\mathbf{T}$$\mathbf{_W}$ $\mathbf{registrados}$ $\mathbf{por}$ $\mathbf{periodo}$",y=0.75)

plt.subplots_adjust(left=None, bottom=0.05, right=None, top=0.85, wspace=0.04, hspace=-0.55)

plt.savefig("des_maximos_MSD.svg",bbox_inches="tight",dpi=300)
plt.savefig("p1_des_maximos.png",bbox_inches="tight",dpi=300)
plt.show()
```
